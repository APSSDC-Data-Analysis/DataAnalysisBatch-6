### 1. Categorical Plot
- By default it returns Scatter Plot
    - sns.catplot()

```python
help(sns.catplot)
```

```python
sns.catplot(x = 'species', y = 'sepal_length', data = iris)
```

```python
sns.catplot(x = 'species', y = 'petal_length', data = iris)
```

#### Categorical Distrubution  Data
- Boxplot

```python
sns.catplot(x = 'species', y = 'sepal_length', data = iris, kind = 'box')
```

```python
sns.catplot(data =iris, kind = 'box') # It takes only numerical data
```

```python
sns.catplot(data =iris, kind = 'box', orient = 'h')
```

```python
sns.catplot(x = 'species', y = 'sepal_width', data = iris, kind = 'box')
```

```python
sns.catplot(x = 'species', y = 'sepal_length', data = iris, kind = 'box', palette = 'husl')
```

- Violin Plot
    - similar to boxplot

```python
sns.catplot(x = 'species', y = 'sepal_width', data = iris, kind = 'violin')
```

### 2. Jiont Plot
- Combination of two plots. By default it have scatterplot, histogram
    - sns.jointplot()

```python
iris.head(2)
```

```python
help(sns.jointplot)
```

```python
sns.jointplot(x = 'petal_width', y = 'sepal_width', data = iris)
```

```python
sns.jointplot(x = 'petal_width', y = 'sepal_width', data = iris, color = 'g')
```

```python
sns.jointplot(x = 'petal_width', y = 'sepal_width', data = iris, kind="hex", color = 'orange')
```

### 3. Pair Plot
- It displays multiple plots at a time in single graph
- Pairwise Relationship
- By default it returns scatterplot
    - sns.pairplot()

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue = 'species')
```

```python
help(sns.pairplot)
```

```python
sns.pairplot(iris, hue="species", palette="husl")
```

```python
sns.pairplot(iris, hue="species", palette="dark")
```

```python
sns.pairplot(iris, hue="species",markers = ['o','D','s'])
```

```python
sns.pairplot(iris, vars = ['sepal_length','sepal_width'])
```

### Heatmap
- It is used to find the correlation two columns in data
    - sns.heatmap()

```python
t = sns.load_dataset('tips')
```

```python
t.head()
```

```python
t.corr()
```

```python
sns.heatmap(t.corr())
```

```python
help(sns.heatmap)
```

```python
sns.heatmap(t.corr(), annot = True)
```

```python
sns.heatmap(t.corr(), annot = True, center = 0, linewidths=5, linecolor='white')
```

```python
sns.heatmap(t.corr(), annot = True, cmap = 'Blues')
```

```python
sns.heatmap(t.corr(), annot = True, cmap='YlGnBu')
```

```python
sns.heatmap(t.corr(), annot = True, cmap = 'RdOr')
```
